## **Getting started**
</div>

The document you are reading is not a static web page, but an interactive environment called a **Colab notebook** that lets you write and execute code.

For example, here is a **code cell** with a short Python script that computes a value, stores it in a variable, and prints the result:

```
seconds_in_a_day = 24 * 60 * 60
seconds_in_a_day
```

To execute the code in the above cell, select it with a click and then either press the play button to the left of the code, or use the keyboard shortcut "Command/Ctrl+Enter". To edit the code, just click the cell and start editing.

Variables that you define in one cell can later be used in other cells:

```
seconds_in_a_week = 7 * seconds_in_a_day
seconds_in_a_week
```

Colab notebooks allow you to combine **executable code** and **rich text** in a single document, along with **images**, **HTML**, **LaTeX** and more. When you create your own Colab notebooks, they are stored in your Google Drive account. You can easily share your Colab notebooks with co-workers or friends, allowing them to comment on your notebooks or even edit them. ## Data science
</div>

With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.

```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

Write a python program to draw a line using given axis values take from a text file in the x axis , y axis and title

```
import matplotlib.pyplot as plt

x_label = 'Values on X-axis'
y_label = 'Values on Y-Axis'
title = 'Line Plot of the Values'

# Define data directly
data = [(1, 2), (2, 3), (3, 5), (4, 7), (5, 11)]  # Example data

x_values = [item[0] for item in data]
y_values = [item[1] for item in data]

plt.plot(x_values, y_values, label='Line Plot', color='r')
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.title(title)
plt.legend()
plt.show()
```

7.Create plot using subplots(), scatter, and some other methods on the Axes object:

```
import matplotlib.pyplot as plt
import numpy as np


data_x, data_y = np.random.rand(50), np.random.rand(50)
bar_data = {'Category A': 20, 'Category B': 35, 'Category C': 15}

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Scatter plot
axs[0, 0].scatter(data_x, data_y, color='green', label='Scatter Plot')

# Line plot
axs[0, 1].plot(data_x, data_y, color='red', linestyle='--', marker='o', label='Line Plot')

# Bar chart
axs[1, 0].bar(bar_data.keys(), bar_data.values(), color='pink', label='Bar Chart', edgecolor='black', width=0.5)

# Histogram
axs[1, 1].hist(np.random.randn(1000), bins=30, color='purple', label='Histogram', edgecolor='black')

# Show the plot
plt.show()
```

8.Generate random data for x and y and create a scatter plot using Matplotlib.

```
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(70)
num_points = 60
x_values = np.random.rand(num_points) * 1.5
y_values = 3 * x_values + 2 + 0.15 * np.random.randn(num_points)

plt.scatter(x_values, y_values, label='Data', color='red')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot of Different Random Data')
plt.legend()
plt.show()
```

11.Create a pie chart to represent the distribution of expenses in a budget. Use fictional data or create your own.

```
import matplotlib.pyplot as plt

categories = ['Housing', 'Utilities', 'Food', 'Entertainment', 'Transportation']
expenses = [1200, 250, 500, 150, 100]
myexplode = [0.02, 0.02, 0.02, 0.01, 0.01]

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(expenses, labels=categories, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'lightcoral', 'gold', 'lightpink'], explode=myexplode)
plt.title('Expense Distribution in Budget')
plt.show()
```
