```python
import numpy as np

def euclidean_distance(x, y):
    return np.linalg.norm(x - y)

def manhattan_distance(x, y):
    return np.sum(np.abs(x - y))

def chebyshev_distance(x, y):
    return np.max(np.abs(x - y))

x = np.array([float(input("Enter x coordinate for point 1: ")), float(input("Enter y coordinate for point 1: ")), float(input("Enter z coordinate for point 1: "))])
y = np.array([float(input("Enter x coordinate for point 2: ")), float(input("Enter y coordinate for point 2: ")), float(input("Enter z coordinate for point 2: "))])

euclidean_dist = euclidean_distance(x, y)
manhattan_dist = manhattan_distance(x, y)
chebyshev_dist = chebyshev_distance(x, y)

print(f"Euclidean Distance: {euclidean_dist}")
print(f"Manhattan Distance: {manhattan_dist}")
print(f"Chebyshev Distance: {chebyshev_dist}")
```

```python
from sklearn.datasets import load_iris
from scipy.spatial.distance import cdist

iris = load_iris()
data = iris.data

distance_matrix = cdist(data, data, 'euclidean')

print("Euclidean Distance Matrix:")
print(distance_matrix)
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
data = iris.data
target = iris.target

df = pd.DataFrame(data, columns=iris.feature_names)
df['target'] = target

class_0 = df[df['target'] == 1]
class_1 = df[df['target'] == 2]

x_attribute = 'sepal length (cm)'
y_attribute = 'sepal width (cm)'

plt.scatter(class_0[x_attribute], class_0[y_attribute], label='Class 0')
plt.scatter(class_1[x_attribute], class_1[y_attribute], label='Class 1')
plt.xlabel(x_attribute)
plt.ylabel(y_attribute)
plt.legend()
plt.show()
```
